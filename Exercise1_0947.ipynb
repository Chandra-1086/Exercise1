```python
import numpy as np
import matplotlib.pyplot as plt

# Hardcoded coefficients
a, b, c = 1, -3, 2

# Define the range for x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Compute the quadratic function
y = a * x**2 + b * x + c

# Plot the quadratic equation
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"{a}xÂ² + {b}x + {c}", color="blue")
plt.title("Quadratic Weather Model (Hardcoded Variables)")
plt.xlabel("x (e.g., time)")
plt.ylabel("y (e.g., temperature)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get coefficients from user input
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Define the range for x values
x = np.linspace(-10, 10, 100)

# Compute the quadratic function
y = a * x**2 + b * x + c

# Plot the quadratic equation
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"{a}xÂ² + {b}x + {c}", color="green")
plt.title("Quadratic Weather Model (Keyboard Input)")
plt.xlabel("x (e.g., time)")
plt.ylabel("y (e.g., temperature)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read coefficients from a file
with open("input.txt", "r") as file:
    line = file.readline()
    a, b, c = map(float, line.split())

# Define the range for x values
x = np.linspace(-10, 10, 100)

# Compute the quadratic function
y = a * x**2 + b * x + c

# Plot the quadratic equation
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"{a}xÂ² + {b}x + {c}", color="red")
plt.title("Quadratic Weather Model (File Input - Single Set)")
plt.xlabel("x (e.g., time)")
plt.ylabel("y (e.g., temperature)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read coefficients from a file
with open("ionput.txt", "r") as file:
    lines = file.readlines()

# Define the range for x values
x = np.linspace(-10, 10, 100)

plt.figure(figsize=(8, 6))

# Loop through each line in the file
for line in lines:
    a, b, c = map(float, line.split())
    y = a * x**2 + b * x + c
    plt.plot(x, y, label=f"{a}xÂ² + {b}x + {c}")

# Plot the quadratic equations
plt.title("Quadratic Weather Models (Multiple Sets from File)")
plt.xlabel("x (e.g., time)")
plt.ylabel("y (e.g., temperature)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```
